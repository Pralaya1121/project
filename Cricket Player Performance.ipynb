```python
mask=delivery['batsman_runs']==4
new_delivery=delivery[mask]
new_delivery.shape[0] #total no of 4 clicked by ipl
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)#total no of 4 clicked by individual player
```

```python
#V Kohil hits maximum number of runs in which team
vk=delivery[delivery['batsman']=='V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
def run_scored (batsman_name):
    vk=delivery[delivery['batsman']=='batsman_name']
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)#missing somthing
```

```python
run_scored('V Kohli')
```

```python
#isin()
#find the most destructive death over batsman in the history of ipl
#srrike rate=(Number of runs)/(Number of balls)*100
#min batsman 200 balls in over 16-20
```

```python
mask=delivery[delivery['over']>15]
all_batsman=mask.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
#all_batsman[x]
batsman_list=all_batsman[x].index.tolist()
```

```python
# Runs scored by all these 43 batsman
# Balls played by these 43 batsman
final=delivery[delivery['batsman'].isin(batsman_list)]
runs=final.groupby('batsman')['batsman_runs'].sum()
balls=final.groupby('batsman')['batsman_runs'].count()
sr=(runs/balls)*100
sr.sort_values(ascending=False)
```
